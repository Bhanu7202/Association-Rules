```python
data = list(df["shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil"].apply(lambda x:x.split(",") ))
data
```

# Transform your transaction data into a one-hot encoded format.

```python

a = TransactionEncoder()  # initialize the encoder
a
```

```python
fit_df = a.fit(data).transform(data)         ## This method  fit () learns the unique items in your DataFrame. ## transform(df) converts the original DataFrame into a one-hot encoded NumPy array
df = pd.DataFrame(fit_df , columns = a.columns_)# converting back to DataFrame
df = df.replace ({False :0, True: 1})
df
```

```python
# lets use a DataFrame with bool type

print(df.dtypes)
```

```python
df = df.astype(bool)
```

#  Apriori Algorithm 

```python
frequent_itemsets = apriori(df, min_support = 0.01, use_colnames = True)     # min_support = 1%
frequent_itemsets.tail(30)
```

```python
#Let's view our interpretation values using the Association rule function.

rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 0.2)
rules
```

```python
rules[rules.lift>1] 

# here If the Lift value is >1 then its treated as ACTIONABLE RULE
# if the lift is bigger then antecedent is given more credit .. 
# for example in 1st row customer is buying mineral water because he\she is buying avacado
# if Lift value is SMALL then antecedent plays no role in that situation the consequent itself is a popular product.

# 2 Itemsets
# since this are actionable insights we can act according like placing avacado near mineral water section OR we can promote 
# mineral water and increase AVACADO prices

# 3 itemsets
# record 431 MINERAL WATER ,SPAGHETTI, PANCAKES (lift = 2) we can place MINERAL WATER ,SPAGHETTI together and place pancakes 
# far at next corner and try to place different items in between them along the path to increase sale ex- AVACADO
```

An leverage value of 0 indicates independence. Range will be [-1 1]Â¶

A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules_confi = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.2)
```

```python
rules_confi

# here we can observe that confidence column
# record 161 has a moderate confidence value which interprets that if customer buys (pancakes, spaghetti) 
# then 45% of people WILL BUY mineral water) as well

#record no 158 if customer is buying (olive oil, mineral water) then theres an 37% chance that he will buy (spaghetti) as well
# in this we can analyse our data
```
